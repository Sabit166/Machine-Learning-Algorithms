```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.columns
```

```python
train = train.drop(['PassengerId', 'Name'], axis = 1)
test = test.drop(['PassengerId', 'Name'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for kalam in train.columns:
    train[kalam] = le.fit_transform(train[kalam])
    
for kalam in test.columns:
    test[kalam] = le.fit_transform(test[kalam])
```

```python
X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
predict = rfc.predict(test)
```

```python
prediction = rfc.predict(test)
y_test = pd.read_csv('sample_submission.csv')
y_test['Transported'] = predict
```

```python
y_test['Transported'] = y_test['Transported'].astype(bool)
```

```python
y_test.to_csv('predictions.csv', index=False)
```

```python
rfc.score(y_test['Transported'], prediction)
```
